```python
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, log_loss, confusion_matrix
from sklearn.model_selection import KFold
from tqdm import tqdm
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import mnist, fashion_mnist, cifar10
from qiskit import Aer, transpile, execute, QuantumCircuit
from qiskit.circuit.library import ZZFeatureMap
from qiskit.circuit import ParameterVector
from qiskit.providers.aer import AerSimulator
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from qiskit.utils import QuantumInstance
from qiskit_machine_learning.kernels import QuantumKernel
import time

# Ensure eager execution is enabled
tf.config.run_functions_eagerly(True)

# Dataset loading and configuration setup
datasets = {
    "MNIST": mnist.load_data(),
    "CIFAR-10": cifar10.load_data()
}

class1 = 0
class2 = 1

n_qubits = 4
n_layers = 6
backend = Aer.get_backend('statevector_simulator')
feature_map = ZZFeatureMap(feature_dimension=n_qubits, reps=1, entanglement='linear')

def preprocess_data(images, labels, class1, class2, img_shape):
    images = images.reshape(-1, img_shape) / 255.0
    scaler = StandardScaler()
    images = scaler.fit_transform(images)
    mask = (labels == class1) | (labels == class2)
    images = images[mask]
    labels = labels[mask]
    labels = np.where(labels == class1, 0, 1)
    return images, labels

def reduce_dataset(images, labels, fraction=0.3):
    size = int(len(images) * fraction)
    indices = np.random.choice(len(images), size, replace=False)
    return images[indices], labels[indices]

def create_quantum_circuit(n_qubits, n_layers, unique_id):
    qc = QuantumCircuit(n_qubits)
    for _ in range(n_layers):
        for i in range(n_qubits):
            theta = ParameterVector(f'θ_{unique_id}_{i}', 1)
            phi = ParameterVector(f'φ_{unique_id}_{i}', 1)
            qc.rx(theta[0], i)
            qc.rz(phi[0], i)
        for i in range(n_qubits - 1):
            qc.cx(i, i + 1)
    return qc

def quantum_circuit_execution(kernel, x, unique_id):
    try:
        circuit = create_quantum_circuit(len(kernel), 1, unique_id)
        param_dict = {param: val for param, val in zip(circuit.parameters, x)}
        bound_circuit = circuit.bind_parameters(param_dict)
        transpiled_circuit = transpile(bound_circuit, backend)
        job = execute(transpiled_circuit, backend)
        while job.status().name not in ['DONE', 'ERROR', 'CANCELLED']:
            time.sleep(1)
        result = job.result()
        statevector = result.get_statevector()
        return np.real(statevector[0])
    except Exception as e:
        print(f'Error during quantum circuit execution: {e}')
        return np.zeros(1, dtype=np.float32)

class QuantumLayer(tf.keras.layers.Layer):
    def __init__(self, n_qubits, n_layers, backend, feature_map, **kwargs):
        super(QuantumLayer, self).__init__(**kwargs)
        self.n_qubits = n_qubits
        self.n_layers = n_layers
        self.backend = backend
        self.feature_map = feature_map
        self.unique_id = 0

    def build(self, input_shape):
        self.kernel = self.add_weight(name='kernel',
                                      shape=(self.n_layers, self.n_qubits, 3),
                                      initializer='uniform',
                                      trainable=True)
        super(QuantumLayer, self).build(input_shape)

    def call(self, inputs):
        def execute_quantum_circuit(inputs, kernel):
            try:
                inputs_np = inputs.numpy()
                result = quantum_circuit_execution(kernel, inputs_np, self.unique_id)
                self.unique_id += 1
                return result
            except Exception as e:
                print(f'Error during quantum circuit execution: {e}')
                return np.zeros(1, dtype=np.float32)

        results = tf.map_fn(lambda x: tf.py_function(execute_quantum_circuit, [x, self.kernel], tf.float32), inputs)
        results = tf.reshape(results, [-1, 1])
        return results

    def compute_output_shape(self, input_shape):
        return (input_shape[0], 1)

def build_hybrid_nn(input_shape, n_qubits, n_layers, backend, feature_map):
    inputs = layers.Input(shape=(input_shape,))
    quantum_output = QuantumLayer(n_qubits, n_layers, backend, feature_map)(inputs)
    classical_output = layers.Dense(64, activation='relu')(quantum_output)
    classical_output = layers.Dense(32, activation='relu')(classical_output)
    outputs = layers.Dense(1, activation='sigmoid')(classical_output)
    model = models.Model(inputs=inputs, outputs=outputs)
    return model

def build_hybrid_nn_with_qpca(input_shape, n_qubits, n_layers, backend, feature_map, n_qpca_components=2):
    inputs = layers.Input(shape=(input_shape,))
    quantum_output = QuantumLayer(n_qubits, n_layers, backend, feature_map)(inputs)
    # Quantum PCA Integration
    pca = PCA(n_components=n_qpca_components)
    quantum_output = pca.fit_transform(quantum_output)
    classical_output = layers.Dense(64, activation='relu')(quantum_output)
    classical_output = layers.Dense(32, activation='relu')(classical_output)
    outputs = layers.Dense(1, activation='sigmoid')(classical_output)
    model = models.Model(inputs=inputs, outputs=outputs)
    return model

def build_regularized_nn(optimizer, input_shape):
    model = models.Sequential()
    model.add(layers.Dense(256, activation='relu', input_shape=(input_shape,), kernel_regularizer=tf.keras.regularizers.l2(0.0001)))
    model.add(layers.BatchNormalization())
    model.add(layers.Dropout(0.3))
    model.add(layers.Dense(128, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.0001)))
    model.add(layers.BatchNormalization())
    model.add(layers.Dropout(0.3))
    model.add(layers.Dense(64, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.0001)))
    model.add(layers.BatchNormalization())
    model.add(layers.Dropout(0.3))
    model.add(layers.Dense(1, activation='sigmoid'))
    model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
    return model

def train_custom_model(model, data, labels, epochs, batch_size):
    optimizer = tf.keras.optimizers.Adam(learning_rate=tf.keras.optimizers.schedules.ExponentialDecay(
        initial_learning_rate=0.01, decay_steps=100, decay_rate=0.96
    ))
    loss_fn = tf.keras.losses.BinaryCrossentropy()
    train_dataset = tf.data.Dataset.from_tensor_slices((data, labels)).batch(batch_size)

    history = {'loss': [], 'val_loss': [], 'accuracy': [], 'val_accuracy': []}
    for epoch in range(epochs):
        print(f'Start of epoch {epoch + 1}')
        epoch_loss = []
        epoch_accuracy = []

        for step, (x_batch, y_batch) in enumerate(train_dataset):
            with tf.GradientTape() as tape:
                logits = model(x_batch, training=True)
                loss_value = loss_fn(tf.expand_dims(y_batch, axis=-1), logits)  # Reshape labels to match logits
                predictions = tf.cast(logits > 0.5, tf.int32)
                y_batch_reshaped = tf.cast(tf.expand_dims(y_batch, axis=-1), tf.int32)
                accuracy = tf.reduce_mean(tf.cast(tf.equal(predictions, y_batch_reshaped), tf.float32))

            grads = tape.gradient(loss_value, model.trainable_weights)
            optimizer.apply_gradients(zip(grads, model.trainable_weights))
            epoch_loss.append(loss_value.numpy())
            epoch_accuracy.append(accuracy.numpy())

            if step % 100 == 0:
                print(f'Step {step}, Loss: {loss_value.numpy()}, Accuracy: {accuracy.numpy()}')

        val_logits = model(data, training=False)
        val_loss = loss_fn(tf.expand_dims(labels, axis=-1), val_logits)  # Reshape labels to match logits
        val_predictions = tf.cast(val_logits > 0.5, tf.int32)
        labels_reshaped = tf.cast(tf.expand_dims(labels, axis=-1), tf.int32)
        val_accuracy = tf.reduce_mean(tf.cast(tf.equal(val_predictions, labels_reshaped), tf.float32))

        history['loss'].append(np.mean(epoch_loss))
        history['val_loss'].append(val_loss.numpy())
        history['accuracy'].append(np.mean(epoch_accuracy))
        history['val_accuracy'].append(val_accuracy.numpy())

        print(f'Epoch {epoch + 1}/{epochs}, Loss: {np.mean(epoch_loss):.4f}, Validation Loss: {val_loss.numpy():.4f}, '
              f'Accuracy: {np.mean(epoch_accuracy):.4f}, Validation Accuracy: {val_accuracy.numpy():.4f}')

    return history

def cross_validate_model(data, labels, input_shape, n_qubits, n_layers, backend, feature_map, k=3, epochs=10, batch_size=32):
    print(f'Data shape: {data.shape}, Labels shape: {labels.shape}')
    kf = KFold(n_splits=k, shuffle=True, random_state=42)
    val_scores = []

    for train_index, val_index in kf.split(data):
        train_data, val_data = data[train_index], data[val_index]
        train_labels, val_labels = labels[train_index], labels[val_index]

        print(f'Train data shape: {train_data.shape}, Train labels shape: {train_labels.shape}')
        print(f'Val data shape: {val_data.shape}, Val labels shape: {val_labels.shape}')

        model = build_hybrid_nn(input_shape, n_qubits, n_layers, backend, feature_map)
        history = train_custom_model(model, train_data, train_labels, epochs, batch_size)
        val_logits = model.predict(val_data)
        val_loss = tf.keras.losses.BinaryCrossentropy()(tf.expand_dims(val_labels, axis=-1), val_logits).numpy()
        val_scores.append(val_loss)

    print(f'Cross-Validation Binary Crossentropy Loss: {np.mean(val_scores)}')

    return history

class QGDOptimizer(tf.keras.optimizers.Optimizer):
    def __init__(self, learning_rate=0.001, name="QGD", **kwargs):
        super().__init__(name, **kwargs)
        self._learning_rate = self._build_learning_rate(learning_rate)

    def _build_learning_rate(self, learning_rate):
        if isinstance(learning_rate, (float, int)):
            return tf.Variable(learning_rate, dtype=tf.float32, trainable=False)
        elif isinstance(learning_rate, tf.Variable):
            return learning_rate
        else:
            raise TypeError("Learning rate should be a float, int, or tf.Variable.")

    def _create_slots(self, var_list):
        for var in var_list:
            self.add_slot(var, "m")

    def _prepare_local(self, var_device, var_dtype, apply_state):
        super()._prepare_local(var_device, var_dtype, apply_state)
        apply_state[(var_device, var_dtype)]["lr_t"] = tf.identity(self._learning_rate)

    def _resource_apply_dense(self, grad, var, apply_state=None):
        var_device, var_dtype = var.device, var.dtype.base_dtype
        coefficients = ((apply_state or {}).get((var_device, var_dtype)) or
                        self._fallback_apply_state(var_device, var_dtype))
        var.assign_sub(coefficients["lr_t"] * grad)

    def _resource_apply_sparse(self, grad, var, indices, apply_state=None):
        raise NotImplementedError("Sparse gradients are not supported.")

    def get_config(self):
        base_config = super().get_config()
        return {
            **base_config,
            "learning_rate": self._learning_rate.numpy(),
        }

    @tf.function
    def update_step(self, gradient, variable):
        lr_t = tf.cast(self._learning_rate, variable.dtype)
        if tf.rank(gradient) == 0:
            gradient = tf.fill(variable.shape, gradient)
        variable.assign_sub(lr_t * gradient)

def train_qgd_model(model, data, labels, epochs, batch_size):
    optimizer = QGDOptimizer(learning_rate=0.01)
    loss_fn = tf.keras.losses.BinaryCrossentropy()
    train_dataset = tf.data.Dataset.from_tensor_slices((data, labels)).shuffle(buffer_size=1024).batch(batch_size)

    history = {'loss': [], 'val_loss': [], 'accuracy': [], 'val_accuracy': []}

    for epoch in range(epochs):
        print(f'Start of epoch {epoch + 1}')
        epoch_loss = []
        epoch_accuracy = []

        with tqdm(total=len(train_dataset), desc=f'Epoch {epoch + 1}/{epochs}', unit='batch') as pbar:
            for step, (x_batch, y_batch) in enumerate(train_dataset):
                with tf.GradientTape() as tape:
                    logits = model(x_batch, training=True)
                    loss_value = loss_fn(tf.expand_dims(y_batch, axis=-1), logits)  # Reshape labels to match logits
                    predictions = tf.cast(logits > 0.5, tf.int32)
                    y_batch_reshaped = tf.cast(tf.expand_dims(y_batch, axis=-1), tf.int32)
                    accuracy = tf.reduce_mean(tf.cast(tf.equal(predictions, y_batch_reshaped), tf.float32))

                grads = tape.gradient(loss_value, model.trainable_weights)
                optimizer.apply_gradients(zip(grads, model.trainable_weights))
                epoch_loss.append(loss_value.numpy())
                epoch_accuracy.append(accuracy.numpy())

                pbar.set_postfix({'Loss': f'{loss_value.numpy():.4f}', 'Accuracy': f'{accuracy.numpy():.4f}'})
                pbar.update(1)

        val_logits = model(data, training=False)
        val_loss = loss_fn(tf.expand_dims(labels, axis=-1), val_logits)  # Reshape labels to match logits
        val_predictions = tf.cast(val_logits > 0.5, tf.int32)
        labels_reshaped = tf.cast(tf.expand_dims(labels, axis=-1), tf.int32)
        val_accuracy = tf.reduce_mean(tf.cast(tf.equal(val_predictions, labels_reshaped), tf.float32))

        history['loss'].append(np.mean(epoch_loss))
        history['val_loss'].append(val_loss.numpy())
        history['accuracy'].append(np.mean(epoch_accuracy))
        history['val_accuracy'].append(val_accuracy.numpy())

        print(f'Epoch {epoch + 1}/{epochs}, Loss: {np.mean(epoch_loss):.4f}, Validation Loss: {val_loss.numpy():.4f}, '
              f'Accuracy: {np.mean(epoch_accuracy):.4f}, Validation Accuracy: {val_accuracy.numpy():.4f}')

    return history



def plot_confusion_matrix(cm, title):
    fig, ax = plt.subplots()
    cax = ax.matshow(cm, cmap='Blues')
    plt.title(title, fontsize=15)
    fig.colorbar(cax)
    plt.xlabel('Predicted', fontsize=12)
    plt.ylabel('True', fontsize=12)
    plt.show()

def plot_history(history, title):
    fig, ax1 = plt.subplots()
    ax1.set_xlabel('Epochs')
    ax1.set_ylabel('Loss', color='tab:blue')
    ax1.plot(history['loss'], label=' Loss', color='tab:blue')
    ax1.plot(history['val_loss'], label='Validation Loss', color='tab:blue', linestyle='dashed')
    ax1.tick_params(axis='y', labelcolor='tab:blue')

    ax2 = ax1.twinx()
    ax2.set_ylabel('Accuracy', color='tab:blue')
    ax2.plot(history['accuracy'], label=' Accuracy', color='tab:blue')
    ax2.tick_params(axis='y', labelcolor='tab:orange')

    fig.tight_layout()
    plt.title(title)
    plt.legend(loc='best')
    plt.show()

def plot_tsne(data, labels, title):
    tsne = TSNE(n_components=2, random_state=42)
    tsne_results = tsne.fit_transform(data)
    plt.figure(figsize=(8, 6))
    plt.scatter(tsne_results[:, 0], tsne_results[:, 1], c=labels, cmap='viridis')
    plt.colorbar()
    plt.title(f't-SNE Visualization - {title}')
    plt.show()

def plot_pca(data, labels, title):
    pca = PCA(n_components=2)
    pca_results = pca.fit_transform(data)
    plt.figure(figsize=(8, 6))
    plt.scatter(pca_results[:, 0], pca_results[:, 1], c=labels, cmap='viridis')
    plt.colorbar()
    plt.title(f'PCA Visualization - {title}')
    plt.show()

def plot_metrics(history, dataset_name):
    plt.figure(figsize=(8, 6))
    plt.plot(history['accuracy'], label='Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.title('Accuracy')
    plt.show()

    plt.figure(figsize=(8, 6))
    plt.plot(history['loss'], label='Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Loss')
    plt.show()


def plot_ground_state_energy(energies, title):
    plt.figure(figsize=(8, 6))
    plt.plot(energies, label='Ground State Energy')
    plt.xlabel('Iteration')
    plt.ylabel('Energy')
    plt.title(f'Ground State Energy - {title}')
    plt.legend()
    plt.show()

def compare_methods_with_qeann_qgd():
    results = {}

    for dataset_name, (train_data, test_data) in datasets.items():
        print(f"Processing {dataset_name}")

        (train_images, train_labels), (test_images, test_labels) = train_data, test_data
        img_shape = np.prod(train_images.shape[1:])
        train_images, train_labels = preprocess_data(train_images, train_labels, class1, class2, img_shape)
        test_images, test_labels = preprocess_data(test_images, test_labels, class1, class2, img_shape)

        # Reduce dataset size to 30%
        train_images, train_labels = reduce_dataset(train_images, train_labels, 0.3)
        test_images, test_labels = reduce_dataset(test_images, test_labels, 0.3)

        input_shape = train_images.shape[1]

        print(f"Training with SGD on {dataset_name}")
        model_sgd = build_regularized_nn(optimizers.SGD(learning_rate=0.001), input_shape)
        history_sgd = model_sgd.fit(train_images, train_labels, epochs=10, validation_data=(test_images, test_labels), batch_size=32, verbose=1)
        sgd_predictions = (model_sgd.predict(test_images) > 0.5).astype("int32")
        sgd_metrics = evaluate_model(sgd_predictions, test_labels)
        cm_sgd = confusion_matrix(test_labels, sgd_predictions)
        plot_confusion_matrix(cm_sgd, f'SGD Confusion Matrix - {dataset_name}')
        plot_metrics(history_sgd.history, dataset_name)

        print(f"Training with Adam on {dataset_name}")
        model_adam = build_regularized_nn(optimizers.Adam(learning_rate=0.001), input_shape)
        history_adam = model_adam.fit(train_images, train_labels, epochs=10, validation_data=(test_images, test_labels), batch_size=32, verbose=1)
        adam_predictions = (model_adam.predict(test_images) > 0.5).astype("int32")
        adam_metrics = evaluate_model(adam_predictions, test_labels)
        cm_adam = confusion_matrix(test_labels, adam_predictions)
        plot_confusion_matrix(cm_adam, f'Adam Confusion Matrix - {dataset_name}')
        plot_metrics(history_adam.history, dataset_name)

        print(f"Training with QEANN-QGD on {dataset_name}")
        model_qgd = build_hybrid_nn_with_qpca(input_shape, n_qubits, n_layers, backend, feature_map)
        history_qgd = train_qgd_model(model_qgd, train_images, train_labels, epochs=10, batch_size=16)
        qgd_predictions = (model_qgd.predict(test_images) > 0.5).astype("int32")
        qgd_metrics = evaluate_model(qgd_predictions, test_labels)
        cm_qgd = confusion_matrix(test_labels, qgd_predictions)
        plot_confusion_matrix(cm_qgd, f'QEANN-QGD Confusion Matrix - {dataset_name}')
        plot_metrics(history_qgd, dataset_name)

        # Plotting ROC Curve
        plot_roc_curve(test_labels, qgd_predictions, f'QEANN-QGD ROC Curve - {dataset_name}')

        # Plotting Ground State Energy
        ground_state_energies = history_qgd['loss']  # Assuming ground state energy is tracked as loss
        plot_ground_state_energy(ground_state_energies, f'Ground State Energy - {dataset_name}')

        results[dataset_name] = {
            'SGD': sgd_metrics,
            'Adam': adam_metrics,
            'QEANN-QGD': qgd_metrics
        }

    return results

compare_results_with_qeann_qgd = compare_methods_with_qeann_qgd()
print(compare_results_with_qeann_qgd)



```
